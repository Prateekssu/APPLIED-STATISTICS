```python
import numpy as np
import matplotlib.pyplot as plt

# Generate binomial data
x = np.random.binomial(n=10, p=0.5, size=10)

# Create a histogram
plt.hist(x, bins=range(12), edgecolor='black', alpha=0.7)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
#assessment3
#create a dictionary with studname,studid,dept,grade
#1.Retrieve using the dept names
#2.Retrieve the iloc of 2
#3.Retrieve the id of the students

import pandas as pd
data = {
    'studname': ['Ankita', 'Anita', 'Ritu', 'Arman','Ashish','Tanu','Ram','Jiban'],
    'studid': [101, 102, 103, 104, 105, 106, 107, 108],
    'dept': ['Computer Science', 'BBA', 'Computer Science', 'Bcom', 'Computer Science', 'BBA', 'Agri', 'Bcom'],
    'grade': ['A', 'B', 'A', 'C','A', 'B', 'D', 'A']
}
df = pd.DataFrame(data)
print(df)

# 1. Retrieve using the dept names
cs_students = df[df['dept'] == 'Computer Science']
print("Students in Computer Science:")
print(cs_students)

# 2. Retrieve the iloc of index 2
iloc_2 = df.iloc[2]
print("Data at index 2:")
print(iloc_2)

# 3. Retrieve the id of the students in Computer Science
cs_student_ids = df[df['dept'] == 'Computer Science']['studid']
print("IDs of students in Computer Science:")
print(cs_student_ids.tolist())

```
